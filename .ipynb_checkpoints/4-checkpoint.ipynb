```python
import pandas as pd
import numpy as np
```

```python
dt=pd.read_csv("/home/suraj/Downloads/ML1/housing.csv")
dt.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
dt.hist(bins=50, figsize=(20,20))
#save_fig("attribute_histogram_plots")
plt.show()
```

```python
dt.fillna(dt.mean(), inplace=True)

dt.shape
dt.info()
```

```python
str1="NEAR BAY"
str2="NEAR OCEAN"
str3="<1H OCEAN"
str4="INLAND"




for i in range(0,len(dt["ocean_proximity"])):
    if dt["ocean_proximity"][i]==str1:
        dt.at[i,"ocean_proximity"]=1
    
    if dt["ocean_proximity"][i]==str2:
        dt.at[i,"ocean_proximity"]=2
    
    if dt["ocean_proximity"][i]==str3:
        dt.at[i,"ocean_proximity"]=3
        
    if dt["ocean_proximity"][i]==str4 or dt["ocean_proximity"][i]=="ISLAND":
        dt.at[i,"ocean_proximity"]=4
        

dt.astype(dtype='float64')
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(dt, test_size=0.2)
```

```python
train_label=train_set.as_matrix(columns=["median_house_value"])
train_data=train_set.drop("median_house_value",axis=1)
mat_train=train_data.as_matrix()

test_label=test_set.as_matrix(columns=["median_house_value"])
test_data=test_set.drop("median_house_value",axis=1)
mat_test=test_data.as_matrix()

mat_train.shape

```

```python
z=np.ones((len(train_label),1),dtype='float64')
mat_train=np.column_stack((mat_train,z))

z=np.ones((len(test_label),1),dtype='float64')
mat_test=np.column_stack((mat_test,z))
                          

```

```python
mat_train
```

```python
t=np.dot(mat_train.transpose(),mat_train)

print (t)
```

```python
t=np.linalg.inv(t)
print (t.shape)
```

```python
t1=np.dot(t,mat_train.transpose())

```

```python
W=np.dot(t1,train_label)
#W=t1*train_label
print (W,W.shape)
```

```python


for i in range (0,len(W)):
    W1[i]=np.asscalar(W[i])*np.asscalar(W[i])
    

```

```python
W1=0.1*W1
W=W+W1
W
```

```python
predictions=np.zeros((len(test_label),))
W=np.matrix(W)
for i in range(0,len(test_label)):
    predictions[i]= (np.asscalar(mat_test[i]*W))
    
predictions=predictions.transpose()
```

```python
from sklearn.metrics import mean_squared_error

lin_mse = mean_squared_error(test_label, predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(test_label, predictions)
lin_mae
```
